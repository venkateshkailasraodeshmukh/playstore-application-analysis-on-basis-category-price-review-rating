# data manipulation and imputation

```python
#fill the null value with appropriate using aggregate function such as mean ,mode or midian
```

```python
def impute_median(series):
    return series.fillna(series.median())
```

```python
googlefile.Rating=googlefile["Rating"].transform(impute_median)
```

```python
#to insure rating column null value check
```

```python
googlefile.isnull().sum()
```

```python
#now rating null value become zero null value
```

```python
print(googlefile["Type"].mode())
print(googlefile["Current Ver"].mode())
print(googlefile["Android Ver"].mode())
```

```python
#filling values in catagorries
```

```python
googlefile["Type"].fillna(str(googlefile['Type'].mode().values[0]), inplace=True)
googlefile["Current Ver"].fillna(str(googlefile['Current Ver'].mode().values[0]), inplace=True)
googlefile["Android Ver"].fillna(str(googlefile['Android Ver'].mode().values[0]), inplace=True)
```

```python
#check again isnull are zero or not
```

```python
googlefile.isnull().sum()
```

```python
#now we manage our data in proper manner but some other columns having problems to analysing that is price which  is not in intiger let convert them also istalls colmn to
```

```python
googlefile['Price']=googlefile['Price'].apply(lambda x: str(x).replace('$','')  if '$' in str(x) else(x)) 
googlefile['Price']=googlefile['Price'].apply(lambda x: float(x))
googlefile['Reviews']=pd.to_numeric(googlefile['Reviews'], errors='coerce')
```

```python
googlefile.head(10)
```

```python
googlefile['Installs']=googlefile['Installs'].apply(lambda x: str(x).replace('+','')  if '+' in str(x) else(x)) 
googlefile['Installs']=googlefile['Installs'].apply(lambda x: str(x).replace(',','')  if ',' in str(x) else(x))
googlefile['Installs']=googlefile['Installs'].apply(lambda x: float(x))
```

# data visualization


```python
group=googlefile.groupby('Category')
a=group['Rating'].agg(np.mean)
b=group['Price'].agg(np.sum)
c=group['Reviews'].agg(np.mean)
print(a)
print(b)
print(c)
```

```python
plt.figure(figsize=(12,5))
plt.plot(a, 'ro')
plt.xticks(rotation=90)
plt.title('Category wise rating')
plt.xlabel('Categories----->')
plt.ylabel('Rating---->')
plt.show()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(b,'ro')
plt.xticks(rotation=90)
plt.title('Category wise pricing')
plt.xlabel('Categories----->')
plt.ylabel('pricing---->')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(c, 'ro')
plt.xticks(rotation=90)
plt.title('Category wise reviews')
plt.xlabel('Categories----->')
plt.ylabel('reviews---->')
plt.show()
plt.show()
```
